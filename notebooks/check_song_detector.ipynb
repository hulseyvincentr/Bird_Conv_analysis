```python
import pandas as pd
import json

def create_dataframe_from_json(json_path):
    # Load the JSON file
    with open(json_path, 'r') as file:
        data = json.load(file)
    
    # Prepare records for DataFrame
    records = []
    for entry in data:
        # Extract filename and song_present for each entry
        filename = entry['filename']
        song_present = entry['song_present']
        
        # Initialize lists for timebins
        onset_timebin_values = []
        offset_timebin_values = []
        
        # Extract onset_timebin and offset_timebin values for each segment
        for segment in entry['segments']:
            if 'onset_timebin' in segment:
                onset_timebin_values.append(segment['onset_timebin'])
            if 'offset_timebin' in segment:
                offset_timebin_values.append(segment['offset_timebin'])
        
        # Append the record as a dictionary
        records.append({
            'filename': filename,
            'song_present': song_present,
            'onset_timebin': onset_timebin_values,
            'offset_timebin': offset_timebin_values
        })
    
    # Convert list of records to a DataFrame
    df = pd.DataFrame(records)
    return df

# Example usage
json_path = '/Users/mirandahulsey-vincent/Documents/allPythonCode/Bird_Conv_analysis/input/onset_offset_results.json'
json_dataframe = create_dataframe_from_json(json_path)

# Display the DataFrame
print("Onset values:")
print(json_dataframe['onset_timebin'])
print("Offset values:")
print(json_dataframe['offset_timebin'])
```

```python
import os
import wave

def get_wav_duration(file_path):
    try:
        with wave.open(file_path, 'r') as wav_file:
            frame_rate = wav_file.getframerate()
            num_frames = wav_file.getnframes()
            duration_seconds = num_frames / frame_rate
            duration_minutes = duration_seconds / 60  # Convert seconds to minutes
        return duration_minutes
    except wave.Error:
        print(f"Skipping file (not a valid WAV): {file_path}")
        return 0

def calculate_total_duration(folder_path):
    total_duration = 0
    wav_files = [f for f in os.listdir(folder_path) if f.endswith('.wav')]
    
    for wav_file in wav_files:
        file_path = os.path.join(folder_path, wav_file)
        duration = get_wav_duration(file_path)
        if duration > 0:
            print(f"File: {wav_file}, Duration: {duration:.2f} minutes")
            total_duration += duration
    
    return total_duration

# Set the folder path in the script
folder_path = '/Volumes/Rose-SSD/conv/WAVs/Canary_Echo/'

```

```python
import os
import wave

def get_wav_info(file_path):
    try:
        with wave.open(file_path, 'r') as wav_file:
            frame_rate = wav_file.getframerate()  # Sampling rate
            num_frames = wav_file.getnframes()
            duration_seconds = num_frames / frame_rate
            duration_minutes = duration_seconds / 60  # Convert seconds to minutes
        return duration_minutes, frame_rate
    except wave.Error:
        print(f"Skipping file (not a valid WAV): {file_path}")
        return 0, None

def calculate_total_duration(folder_path):
    total_duration = 0
    wav_files = [f for f in os.listdir(folder_path) if f.endswith('.wav')]
    
    for wav_file in wav_files:
        file_path = os.path.join(folder_path, wav_file)
        duration, frame_rate = get_wav_info(file_path)
        if duration > 0:
            print(f"File: {wav_file}, Duration: {duration:.2f} minutes, Sampling Rate: {frame_rate} Hz")
            total_duration += duration
    
    return total_duration

# Set the folder path in the script
folder_path = '/Volumes/Rose-SSD/conv/WAVs/Canary_Echo/'

# Calculate total duration
total_duration = calculate_total_duration(folder_path)
print(f"\nTotal duration of all valid .wav files: {total_duration:.2f} minutes")
```

```python
import os
import wave
import numpy as np
import pandas as pd

def get_wav_info(file_path):
    try:
        with wave.open(file_path, 'r') as wav_file:
            frame_rate = wav_file.getframerate()  # Sampling rate
            num_frames = wav_file.getnframes()
            duration_seconds = num_frames / frame_rate
        return duration_seconds, frame_rate
    except wave.Error:
        print(f"Skipping file (not a valid WAV): {file_path}")
        return None, None

def generate_song_vector(file_path):
    duration, frame_rate = get_wav_info(file_path)
    if duration is not None and frame_rate is not None:
        num_time_windows = int(frame_rate * duration)
        song_vector = np.zeros(num_time_windows)
        return song_vector, duration, frame_rate
    else:
        return None, None, None

def process_folder(folder_path):
    records = []
    wav_files = [f for f in os.listdir(folder_path) if f.endswith('.wav')]
    
    for wav_file in wav_files:
        file_path = os.path.join(folder_path, wav_file)
        song_vector, duration, frame_rate = generate_song_vector(file_path)
        
        if song_vector is not None:
            records.append({
                'filename': wav_file,
                'duration_minutes': duration / 60,  # Convert seconds to minutes
                'sampling_rate': frame_rate,
                'song_vector': song_vector
            })
    
    # Convert list of records to DataFrame
    df = pd.DataFrame(records)
    return df

# Set the folder path in the script
folder_path = '/Volumes/Rose-SSD/conv/WAVs/Canary_Echo/'

# Process the folder and get the DataFrame
wav_files_and_song_vector_dataframe = process_folder(folder_path)

# Display the DataFrame
print(wav_files_and_song_vector_dataframe.head(2))
```

## For each file in the folder, check the .json file, then set the song_vector points using the onset and offset values for the timebins to 1

```python
import pandas as pd

# Assuming wav_files_and_song_vector_dataframe is your main DataFrame
# and json_dataframe is the secondary DataFrame with onset and offset timebins

# Initialize new columns in wav_files_and_song_vector_dataframe to NaN
wav_files_and_song_vector_dataframe['onset_timebin'] = None
wav_files_and_song_vector_dataframe['offset_timebin'] = None

# Iterate through each row in wav_files_and_song_vector_dataframe
for index, row in wav_files_and_song_vector_dataframe.iterrows():
    filename = row['filename']
    # Find the matching row in json_dataframe
    matching_row = json_dataframe[json_dataframe['filename'] == filename]
    
    if not matching_row.empty:
        # Get the onset and offset timebins from json_dataframe
        onset_timebin = matching_row.iloc[0]['onset_timebin']
        offset_timebin = matching_row.iloc[0].get('offset_timebin', None)
        
        # Update the DataFrame with these values
        wav_files_and_song_vector_dataframe.at[index, 'onset_timebin'] = onset_timebin
        wav_files_and_song_vector_dataframe.at[index, 'offset_timebin'] = offset_timebin

# Display the updated DataFrame
print(wav_files_and_song_vector_dataframe.head(1))
```
